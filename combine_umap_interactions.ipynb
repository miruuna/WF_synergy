```python
import pandas as pd
```

```python
df_interactions = pd.read_csv("WF1a_WF2_interactions_true_false.csv")
df_interactions[["Peptide1", "Peptide2"]] = df_interactions["Peptide_Pair"].str.split("_", expand=True)
df_interactions = df_interactions.drop(columns=["Peptide_Pair"])
df_interactions
```

```python
df_interactions_pep5 = df_interactions.loc[(df_interactions["Peptide1"].isin(["pep5"])) | (df_interactions["Peptide2"].isin(["pep5"]))]
df_interactions_pep5["interacting_peptide"] = df_interactions_pep5.apply(lambda row: row.loc["Peptide1"] if row.loc["Peptide1"]!="pep5" else row.loc["Peptide2"], axis=1)
df_interactions_pep5 = df_interactions_pep5.drop(columns=["Peptide1", "Peptide2"])
df_interactions_pep5 = df_interactions_pep5.set_index("interacting_peptide")
df_interactions_pep5.columns = [float(x) for x in df_interactions_pep5.columns.tolist()]
# df_interactions_pep5.loc["interacting"] = [True if 1 in df_interactions_pep5[x].values else False for x in df_interactions_pep5.columns.tolist()]
df_interactions_pep5.loc["interacting"] = [(df_interactions_pep5[x]==1).sum()for x in df_interactions_pep5.columns.tolist()]

df_interactions_pep5
```
